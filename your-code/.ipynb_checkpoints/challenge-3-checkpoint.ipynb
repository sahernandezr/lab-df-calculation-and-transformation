## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
#Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?
customers = orders.groupby(['CustomerID']).agg({'amount_spent':sum})
customers
```

```python
#Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?
#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?
#First, identify two groups of customers from the data set. The first group is VIP Customers whose aggregated 
#expenses at your global chain stores are above the 95th percentile (aka. 0.95 quantile). 
#The second group is Preferred Customers whose aggregated expenses are between the 75th and 95th percentile.
q95 = customers['amount_spent'].quantile(0.95)
#q95 is 5840.181999999983
q75 = customers['amount_spent'].quantile(0.75)
#q75 is 1661.64
q100 = customers['amount_spent'].quantile(1)
# q100 is 280206.01999999984 = customers['amount_spent'].max()
```

```python
customers['Label']=''

conditions = [
    (customers['amount_spent']>q95),
    (customers['amount_spent']>=q75)
]

labels = ['VIP', 'Preferred']

customers['Label'] = np.select(conditions, labels)

customers
```

# Q2: How to identify which country has the most VIP Customers?

Provide your solution for Q2 below:

```python
total_spent=orders.groupby(["CustomerID","Country"])["amount_spent"].sum().sort_values(ascending=False)
total_spent=total_spent.to_frame()
total_spent['Label']=''

conditions = [
    (total_spent['amount_spent']>q95),
    (total_spent['amount_spent']>=q75)
]

labels = ['VIP', 'Preferred']

total_spent['Label'] = np.select(conditions, labels)

total_spent
```

```python
total_spent[total_spent["Label"]=="VIP"].groupby("Country").count()
#United Kingdom has most of the VIP customers
```

# Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
options = ['VIP', 'Preferred']

VIP_Preferred = total_spent[total_spent['Label'].isin(options)]
VIP_Preferred.groupby("Country").count()
```
